```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.drop(columns=['Purchased'],inplace=True)
```

```python
df
```

```python
def categorize_age(age):
  if age<30:
    return 'Young'
  elif 30<=age<60:
    return 'Middle-aged'
  else:
    return 'Old'
df['Age_Group']=df['Age'].apply(categorize_age)
```

```python
df
```

```python
summary_stats=df.groupby('Age_Group')['EstimatedSalary'].describe()
```

```python
summary_stats
```

```python
def calculate_mean(group):
  return group['EstimatedSalary'].mean()
df.groupby('Age_Group').apply(calculate_mean)
```

```python
def calculate_median(group):
  return group['EstimatedSalary'].median()
df.groupby('Age_Group').apply(calculate_median)
```

```python
def calculate_mode(group):
  return group['EstimatedSalary'].mode()
df.groupby('Age_Group').apply(calculate_mode)
```

```python
def calculate_min(group):
  return group['EstimatedSalary'].min()
df.groupby('Age_Group').apply(calculate_min)
```

```python
def calculate_max(group):
  return group['EstimatedSalary'].max()
df.groupby('Age_Group').apply(calculate_max)
```

```python
def calculate_std(group):
  return group['EstimatedSalary'].std()
df.groupby('Age_Group').apply(calculate_std)
```
