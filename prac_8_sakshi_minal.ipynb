```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset.replace(np.nan, dataset['age'].mean, inplace=True)
dataset.replace(np.nan, dataset['embark_town'].mean, inplace=True)
dataset.replace(np.nan, dataset['embarked'].mean, inplace=True)
dataset['deck'].fillna(dataset['deck'].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['fare'])
```

```python
sns.distplot(dataset['fare'], kde=False)
```

```python
sns.histplot(dataset['fare'], kde=False)
```

```python
sns.histplot(dataset['fare'], kde=False, bins=10)
```
